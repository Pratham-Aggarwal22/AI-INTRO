```python
import warnings
warnings.filterwarnings(action = 'ignore')


from transformers import pipeline
from transformers.utils import logging
logging.set_verbosity(40)

from IPython.display import Image
```

```python
#!gdown 1-NjCVShA4TtsaP8y-XT46EKiUimokxgF
File2 = "Image2.jpg"
```

```python
#!gdown 1-XFemhLburUo9TjsHB33yTdvPYV5tu5D
File3 = "Image3.jpg"
```

```python
Transformer1 = pipeline(task="image-to-text",
                     model="nlpconnect/vit-gpt2-image-captioning")


Transformer2 = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")

Transformer3 = pipeline("image-to-text", model="Salesforce/blip2-opt-2.7b")
```

# Using GPT

```python
print("Image1")
Transformer1(File1)
```

```python
print("Image2")
Transformer1(File2)
```

```python
print("Image3")
Transformer1(File3)
```
